### Problem 3

```python
#sensitivity P(+|M)
sens = 0.917

#specificity P(-|M across)
spec = 0.935

#formula for positive test
def post_pos(prior): 
  post = (sens*prior) / (sens*prior + (1-spec)*(1-prior))
  return(post)


# formula for negative test
def post_neg(prior):
  post = (1-sens)*prior / ((1-sens)*prior + spec*(1-prior))
  return(post)
```

```python
post_neg(post_pos(0.03))
```

```python
post_pos(post_neg(0.03))
```

```python
post_neg(post_neg(0.03))
```

### Problem 6

```python
y = np.array([0, 1, 2, 3, 4, 5, 4, 3, 2, 1, 0])
prior = y / np.sum(y)
prior
```

```python
x = np.arange(0, 11)
plt.bar(x, prior, width=0.3)
```

```python
like = np.arange(0, 1.1, 0.1)
like
```

```python
plt.bar(x, like, width=0.3)
```

```python
margin = np.sum(prior * like)
post = prior * like / margin
post
```

```python
plt.bar(x, post, width=0.3)
```

```python
prior_H = post
margin = np.sum(prior_H * like)
post_H = prior_H * like / margin
np.round(post_H, 3)
```

```python
plt.bar(x, post_H, width=0.3)
```

```python
like = np.arange(0, 1.1, 0.1)
like_T = 1 - like
like_T
```

```python
plt.bar(x, like_T, width=0.3)
```

```python
prior_T = post
margin = np.sum(prior_T * like_T)
post_T = prior_T * like_T / margin
np.round(post_T, 3)
```

```python
plt.bar(x, post_T, width=0.3)
```

```python
plt.subplots(1,2, sharey="all")

plt.subplot(121)
plt.bar(x, prior, width=0.3)

plt.subplot(122)
plt.bar(x, post_T, width=0.3)
```
